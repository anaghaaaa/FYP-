Load the Wisconsin Breast Cancer Diagnostic dataset (wdbc.csv) into a new DataFrame called data while specifying the header and column names. The column_names list is constructed to match the structure of the dataset, with 'id', 'diagnosis', and feature columns.

```python
# Load the Wisconsin Breast Cancer Diagnostic dataset into a DataFrame
column_names = ['id', 'diagnosis'] + [f'feature_{i}' for i in range(1, 31)]
data = pd.read_csv('/content/wdbc.csv',header=None, names=column_names)
```

Separate the features (X) and the target variable (y) from the data DataFrame. X contains the feature columns (excluding 'id' and 'diagnosis'), and y contains the 'diagnosis' column.

```python
# Separate features (X) and target variable (y)
X = data.drop(columns=['id', 'diagnosis'])
y = data['diagnosis']

```

Apply one-hot encoding to the feature columns in the X DataFrame using pd.get_dummies(). This converts categorical variables into binary columns (dummy variables) to prepare the data for the model.

```python

# Apply one-hot encoding to the feature columns
X_encoded = pd.get_dummies(X)
```

Split the data into training and testing sets using the train_test_split() function. X_train and y_train represent the training data, while X_test and y_test represent the testing data. The test_size parameter determines the proportion of data used for testing (20% in this case).

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

Define a function str_to_float that converts the diagnosis labels ('M' for malignant and 'B' for benign) to float values (1.0 for malignant and 0.0 for benign). Apply this function to the y_train Series to create y_train_floats.

```python
def str_to_float(s):
    if s == 'M':
        return 1.0
    elif s == 'B':
        return 0.0
    else:
        raise ValueError('Could not convert string to float: {}'.format(s))


y_train_floats = y_train.apply(str_to_float)
```

Define the sigmoid activation function and its derivative. The sigmoid function is used to introduce non-linearity to the neural network, and its derivative is used during backpropagation for weight updates.

```python
# Define sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x.astype(float)))

def sigmoid_derivative(x):
    return x * (1 - x)
```

Define the neural network architecture. In this case, the network has input_size input neurons, hidden_size hidden neurons, and output_size output neurons.

```python
# Define neural network architecture
input_size = X_train.shape[1]
hidden_size = 1
output_size = 1
```

Initialize the weights for the input-hidden and hidden-output layers of the neural network with random values between -1 and 1 using np.random.uniform()

```python
# Initialize random weights for input-hidden and hidden-output layers
np.random.seed(0)
weights_input_hidden = np.random.uniform(-1, 1, (input_size, hidden_size))
weights_hidden_output = np.random.uniform(-1, 1, (hidden_size, output_size))
```

Set hyperparameters for the training process: learning_rate controls the step size in weight updates, and epochs determines the number of training iterations.

```python
# Hyperparameters
learning_rate = 0.01
epochs = 5
```

Reshape the training data to match the expected input format of the neural network. The data is reshaped to a 3D array with dimensions (batch_size, 1, num_features).

```python
# Reshape training data to match the network's input format
X_train = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
```

Perform the training loop for the specified number of epochs. In each epoch, the code performs a forward pass through the network to calculate predictions, calculates the error

```python
# Training loop
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")
    # Forward pass
    hidden_input = np.dot(X_train, weights_input_hidden)
    hidden_output = sigmoid(hidden_input)
    output = np.dot(hidden_output, weights_hidden_output)
    output = sigmoid(output)

    # Calculate error and deltas for backpropagation
    error = y_train_floats.values.reshape(-1, 1) - output
    d_output = error * sigmoid_derivative(output)
    d_hidden_output = d_output.dot(weights_hidden_output.T) * sigmoid_derivative(hidden_output)

    # Print error for each iteration within the epoch
    for iteration, err in enumerate(np.mean(np.abs(error), axis=1)):
        print(f"Iteration {iteration + 1}/{X_train.shape[0]} - Error: {err[0]:.6f}")  # Notice the [0]
```

Make predictions on the test set using the trained model.

```python
# Make predictions on the test set
hidden_input_test = np.dot(X_test, weights_input_hidden)
hidden_output_test = sigmoid(hidden_input_test)
predictions = np.dot(hidden_output_test, weights_hidden_output)
predictions = sigmoid(predictions)
```

Round the predictions to binary values using a threshold of 0.5.

```python
# Round predictions to binary values
rounded_predictions = (predictions > 0.5).astype(int)
```

 Calculate the confusion matrix using the confusion_matrix() function and print it. Calculate and print the accuracy using the confusion matrix.

```python
# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test.apply(str_to_float), rounded_predictions)
print("Confusion Matrix:")
print(conf_matrix)
```
